Logic, Loops and Iterations

Simple loop for summation

```python
import numpy as np

a = 0
for j in range(0, 5):
    a = a + (j+1)    
    display(a)
    
```

Simple loop for summation stepping by 2

```python
a = 0
for j in range(0,5,2):
    a = a + (j+1)    
    display(a)
    
```

Simple loop for summation over defined set of numbers

```python
a = 0
loop = [1,5,4]
for j in loop:
    a = a + j 
    display(a)
    
    
```

Bisection example

```python
import matplotlib.pyplot as plt

dx = 0.1
x = np.arange(-10,10+dx,dx)
y = np.exp(x)-np.tan(x)
y2 = 0*y # plot the y=0 line

plt.plot(x,y)
plt.plot(x,y2)
plt.axis([-10, 10, -10, 20])
```

Algorithm for bisection

```python
xr = -2.8; xl = -4
for j in range(0, 100):
    xc = (xr + xl)/2
    fc = np.exp(xc) - np.tan(xc)   
    if ( fc > 0 ):
        xl = xc   
    else:
        xr = xc
        
    if ( abs(fc) < 1e-5 ):        
        display(xc); display(j)
        break
```
